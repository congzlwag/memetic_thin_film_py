```python
import numpy as np
from matplotlib import pyplot as plt

import sys
sys.path.append('../algorithm/')
```

```python
from DispData import DispData
from refl_disp import refl_disp_norm
```

```python
wl_sample_fine = np.array([2.5, 22, 1000])

dipode = DispData(["HfO2","MgF2","SiC","Ag"], wl_sample_fine, wl_sample_fine)
```

```python
Yudisp = {}
for mat, dat in zip(["HfO2","MgF2","SiC","Ag"], dipode.ref_ind_fine.T):
    Yudisp[mat] = dat
```

```python
dipode.disp_plot('Ag')
```

```python
thicknesses = np.array([3.279, 0.640, 0.657, 1.940, 1.337, 0.047])
n_middle = np.array([Yudisp['MgF2'], Yudisp['HfO2'], Yudisp['SiC'], Yudisp['MgF2'], Yudisp["SiC"], Yudisp['MgF2']]).T
```

```python
R_norm = refl_disp_norm(n_in=1, n_disp=n_middle, n_out=Yudisp['Ag'], d=thicknesses, wvlen=dipode.wvlen_vec_fine)
```

```python
plt.plot(dipode.wvlen_vec_fine, 1- R_norm)
plt.xlabel('wavelength ($\mu$m)')
plt.ylabel('emissitivity')
```
